# Weather Data in Seattle from 1948 to 2017

```python
# Global installs for required packages
import pandas as pd
import numpy as np
from datetime import date
```

```python
# read in the weather data file with pandas
df = pd.read_csv('./seattleWeather_1948-2017.csv')
df.head()



df.filter?
```

```python
# Reset this column to have datetime object types
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
```

```python
# Create new columns for data parsing by date information
df['DAY'] = df['DATE'].dt.weekday_name
df['DAY_OF_THE_WEEK'] = df['DATE'].apply(date.isoweekday)
df['MONTH'] = df['DATE'].dt.month
df['YEAR'] = df['DATE'].dt.year
# Change the Rain column to 0/1s rather than True/False
df['RAIN'] = df['RAIN'].apply(lambda x: 1 if x is True else 0)
```

```python
df.head()
```

# 1. How many days in the total dataset has it rained vs not?

```python
pd.DataFrame(df['RAIN'].groupby(df.RAIN).count())
```

# 2. What's the count of rainy days by day of the week?

```python
rain = df[df.RAIN == True].groupby(df.DAY)
rain['DAY'].count()
```

```python
months_wet = df[df.RAIN == True].groupby(df.MONTH)
months_wet['MONTH'].count()
```

# 3. What's the total rainfall per year

```python
prcp_yearly = df[['YEAR', 'PRCP']].groupby(df.YEAR)
sum_prcp = pd.DataFrame(prcp_yearly['PRCP'].sum())
sum_prcp.head()
```

```python
sum_prcp.sort_values('PRCP')
```
